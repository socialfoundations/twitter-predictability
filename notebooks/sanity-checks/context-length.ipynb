```python
from prompting import user_nlls
from collections import defaultdict
from metrics import torch_compute_confidence_interval
from tqdm import tqdm
```

```python
user_id = "1308026329"
modes = ["none", "user", "peer", "random"]

config = {
    "from_disk": True,
    "device": "cuda",
    "user_id": user_id,
    "model_id": "gpt2",
#     "ctxt_len": 900,
#     "window_len": None,
    "stride": "half",
#     "mode": "none",
    "seq_sep": "\n",
    "batched": True,
    "batch_size": 8,
    "token_level_nlls": True,
}
```

## Rerun

### Fix window length to 124, vary context length

```python
context_lengths = [0, 150, 300, 450, 600, 750, 900]
config["window_len"] = 1024 - context_lengths[-1] # 124
```

```python
context_len_results = []

for cl in tqdm(context_lengths, position=0):
    config["ctxt_len"] = cl
    for mode in tqdm(modes, position=1, leave=False):
        config["mode"] = mode
        nlls = user_nlls(config)
        mean, ci = torch_compute_confidence_interval(data=nlls, confidence=0.9)
        
        context_len_results.append({
            "mean": mean,
            "ci": ci,
            "context": mode,
            "l_ctxt": cl
        })
```

### Fix  context length to 450, vary window_length

```python
1024 - 450
```

```python
window_lengths = [50, 100, 150, 250, 350, 450, 574]
config["ctxt_len"] = 1024 - window_lengths[-1] # 450
```

```python
window_len_results = []

for wl in tqdm(window_lengths, position=0):
    config["window_len"] = wl
    for mode in tqdm(modes, position=1, leave=False):
        config["mode"] = mode
        nlls = user_nlls(config)
        mean, ci = torch_compute_confidence_interval(data=nlls, confidence=0.9)
        
        window_len_results.append({
            "mean": mean,
            "ci": ci,
            "context": mode,
            "l_win": wl
        })
```

### Vary BOTH context length, and window length

```python
# setting only context length, and window length to None
context_lengths = [0, 150, 300, 450, 600, 750, 900, 1000]
config["window_len"] = None # this will automatically set it to 1024 - ctxt_len

results = []

for cl in tqdm(context_lengths, position=0):
    config["ctxt_len"] = cl
    for mode in tqdm(modes, position=1, leave=False):
        config["mode"] = mode
        nlls = user_nlls(config)
        mean, ci = torch_compute_confidence_interval(data=nlls, confidence=0.9)
        
        results.append({
            "mean": mean,
            "ci": ci,
            "context": mode,
            "l_ctxt": cl
        })
```
